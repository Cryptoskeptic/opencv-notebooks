Feature extraction using SURF

### Imports

```python
from matplotlib import pyplot as plt

import cv2
import numpy as np

print('OpenCV version: {}'.format(cv2.__version__))
```

### Read image

```python
image = cv2.imread(r"mariposa.jpg")
print('size: {} shape: {}'.format(image.size, image.shape))
```

### Convert to gray scale

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

### Create SURF object
Setup threashold to 400

```python
surf = cv2.xfeatures2d.SURF_create(400)
```

### Find keypoints and descriptor

```python
keypoints, descriptor = surf.detectAndCompute(gray, None)
print('keypoint len: {}'.format(len(keypoints)))
```

### Show image

```python
result = cv2.drawKeypoints(gray, keypoints, image, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(result)
plt.show()
```

### Reduce number of keypoints

```python
print('Hessian threshold: {}'.format(surf.getHessianThreshold()))

# Update the hessian threshold (only for image representation)
surf.setHessianThreshold(50000)

# Get again keypoints and descriptor
keypoints, descriptor = surf.detectAndCompute(gray, None)
print('keypoint len: {}'.format(len(keypoints)))
```

### Draw final
All keypoints are represented in red color

```python
result = cv2.drawKeypoints(gray, keypoints, image, (255, 0, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(result)
plt.show()
```

```python
cv2.imwrite('surf-result.jpg', result)
```
