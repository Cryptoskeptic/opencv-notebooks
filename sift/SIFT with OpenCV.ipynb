Feature extraction using SIFT

### Imports


```python
from matplotlib import pyplot as plt
import cv2
import numpy as np

print('OpenCV version: {}'.format(cv2.__version__))
```

### Read image
Read the image and print basic metadata like syze, number of cols, rows and channels

```python
image = cv2.imread(r"santiago.jpg")
print('size: {} shape: {}'.format(image.size, image.shape))
```

### Convert image to gray scale

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

### Keypoint detection
First create the SIFT object. Calling `sift.detect()` method finds the keypoints in the image

```python
sift = cv2.xfeatures2d.SIFT_create()
keypoints = sift.detect(gray, None)
```

### Visualize the points 
Using the flag `cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS` will draw a circle with size of the keypoint and it will even show its orientation

```python
result = cv2.drawKeypoints(gray, keypoints, image, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
cv2.imwrite('city_sift.jpg', result)

plt.imshow(result)
plt.show()
```

### Calculate descriptor

```python
keypoints, descriptor = sift.compute(gray, keypoints)
print('descriptor: {}'.format(len(descriptor)))
```
